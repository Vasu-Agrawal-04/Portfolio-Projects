```python
#importing necessary libraries
import pandas as pd
import json as json
```

```python
with open("C:\\Users\\vasua\\OneDrive\\Desktop\\DATA ANALYTICS\\PROJECT - CRICKET ANALYSIS\\Prerequisites\\data-zip\\t20_wc_match_results.json", encoding='utf-8', errors='ignore') as f:
    data = json.load(f)
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match_id'},axis = 1,inplace = True)
df_match.head()
```

```python
match_ids_dict = {}
for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']

    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
match_ids_dict
```

#batting summary data

```python
with open("C:\\Users\\vasua\\OneDrive\\Desktop\\DATA ANALYTICS\\PROJECT - CRICKET ANALYSIS\\Prerequisites\\data-zip\\t20_wc_batting_summary.json", encoding='utf-8', errors='ignore') as f:
    data = json.load(f)

    all_rec = []
    for rec in data:
        all_rec.extend(rec['battingSummary'])
df_batting = pd.DataFrame(all_rec)
df_batting.head(15)
```

```python
df_batting["out/not out"] = df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not out")
df_batting.head(15)
```

```python
# df_batting.drop(columns=['dismissal'],inplace = True)
df_batting.head(20)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x : x.replace('†',''))
df_batting.head(20)
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head(20)
```

```python
# Save the DataFrame to CSV with UTF-8 encoding
df_batting.to_csv("C:\\Users\\vasua\\OneDrive\\Desktop\\DATA ANALYTICS\\PROJECT - CRICKET ANALYSIS\\Python - code (Pandas)\\cricket_batting_data.csv", 
                  index=False, 
                  encoding='utf-8')

print("File saved successfully!")
```
